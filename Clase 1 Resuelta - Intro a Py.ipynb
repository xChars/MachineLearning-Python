## Estructuras de control

`Python` a diferencia de `R` trabaja con indentación en vez de parentesis de llave para encapsular lo que está dentro de una estructura de control (ciclo, función, etc). La identación es un tabulador o 4 espacios sucesivos a la linea siguiente (enter) de la estructura de control. Es decir, algo que en `R` sería de la siguiente forma:

```r
if (condición) {
    acción
}
```
En `python` es:

```python
if condicion:
    acción
```
**¡Notar los 4 espacios debajo del if!**

````python
nota = 3.5

if nota > 5.5:
    print("Eximido")
elif nota > 4.0:
    print("A dar examen")
else:
    print("A Salvar el ramo")
````

Los ciclos en `python` se caracterizan por tener una sintaxis un poco distinta, aunque intuitiva cuando uno se acostumbra. Tradicionalmente, uno definir un índice que recorre un conjunto de valores `for(i=0, i<n, i++)`. Python recibe 
- Una lista de valores sobre los que itera
- El nombre de una variable que irá recorriendo esa lista
- Las instrucciones que se deben ejecutar

````python
print("Meses:", meses)

for mes in meses:
    print("Mes:", mes)
````

### Listas por comprensión

Su principal objetivo es iterar sobre muchos valores de una lista, de manera más ordenada y corta. Buscaremos convertir el siguiente código:

````python
colores = ["ROJO", "VERDE", "AZUL", "AMARILLO", "NARANJO"]

lista_auxiliar = []
for color in colores:
    lista_auxiliar.append(color.startswith("A"))

lista_auxiliar
````

Ahora con listas por comprensión:

````python
comienza_con_A = [color.startswith("A") for color in colores]
comienza_con_A
````

Y si necesitamos ver solamente los colores que comiencen con "A", lo que es un filtro:

````python
colores_A = [color for color in colores if color.startswith("A")]
colores_A
````

## Iteradores clásicos

Los iteradores son objetos cuyos valores aún no son generados, o sea contienen la información necesaria para producirlos cuando se utilice una estructura `for` o se le aplique la función `list` en el futuro. Esto es útil en términos de memoria. Por ejemplo, cuando cargamos 600,000 imágenes y aplicamos un filtro de colores, es conveniente cargar la foto, aplicar filtro, guardarla con el nuevo filtro a la vez y no todas al mismo tiempo.
Una vez que iteremos sobre el iterador, debemos definirlo nuevamente.

Algunas funciones que retornan iteradores son:

* `range(i, f, s)`: Crea una secuencia de datos desde `i` hasta `f` (este último no inclusivo con un salto `s`).
* `enumerate(lista)`: Retorna un iterador compuesta por tuplas que contienen el par `posición`, `elemento`.
* `zip(lista1, lista2)`: Retorna un iterador compuesto por tuplas que contienen elementos en la misma posición.

**Ejemplo range**


````python
# Secuencia de valores desde el 2 hasta el 8 de salto 2.
secuencia = range(2, 9, 2)   
print(secuencia)
````

El resultado es un iterador: para ver los valores, debemos aplicar la función `list` o un ciclo `for`. 

````python
print(list(secuencia))
````

````python
secuencia = range(2, 9, 2)
for i in secuencia:
    print("Valor:", i)
````

**Ejemplo enumerate**

````python
numero_y_mes = enumerate(meses)
list(numero_y_mes)
````

````python
for num, mes in enumerate(meses): # destructuración
    print("Mes", num + 1, "es", mes)
````

**Ejemplo zip**

````python
mes_y_poleras = zip(meses, poleras)
list(mes_y_poleras)
````

## Creación de funciones

Las funciones en python trabajan de manera similar a `R` a través del comando `def`

````python
def fahrenheit_a_celcius(temperatura):
  calculo = (temperatura - 32) / 1.8
  return calculo
````

````python
fahrenheit_a_celcius(3)
````

Tampoco es necesario definir variables auxiliares:

````python
def fahrenheit_a_celcius2(temperatura):
  return (temperatura - 32) / 1.8

fahrenheit_a_celcius2(3)
````

Podemos usar funciones anónimas (o sea que definir su nombre es opcional). Su uso es dentro de otras estructuras con el fin usarlas de forma temporal, comunmente en la librería para análisis de datos `Pandas`.
Su estructura se define como:
```python
lambda parametro: función en terminos de parametro
```


````python
fahrenheit_a_celcius3 = lambda temp: (temp - 32) / 1.8
fahrenheit_a_celcius3(3)
````

**Ejemplo de aplicar una función dentro de listas por comprensión.**

Definir 

$$
f(x) = \left\{
\begin{array}{ll}
2x + 1  & x \leq 0  \\ x^2   & x > 0
\end{array} \right.
$$

````python
f = lambda x: 2 * x + 1 if x <= 0 else x ** 2
valores = [4, -2, 9, 56, -67]
# aplicar a todos los valores
print([f(valor) for valor in valores])
# aplicar solo a los valores mayores a 5, en caso contrario no hacer nada
print([f(valor) if valor > 5 else valor for valor in valores])
````

Otra forma de seleccionar datos es usando operaciones lógicas, mediante las que podemos extraer solo los datos que cumplan ciertos requisitos. Por ejemplo: creamos una población normalmente distribuida de estaturas, con media $\mu=160$ cm y desviación estándar $\sigma=5$ cm.

````python
np.random.seed(2022)  # como set.seed de R, para reproducibilidad en NumPy
estaturas = np.random.normal(loc=160, scale=5, size=10000)
estaturas
````

````python
cond1 = estaturas > 159
cond2 = estaturas < 161

print(estaturas[(estaturas >= 159) & (estaturas <= 161)]) # Otros operadores: distinto != , no ~, o |
print(estaturas[cond1 & cond2])  ## Estas dos cosas son lo mismo
````

**Estadística descriptiva**

````python
print("Promedio    :", np.mean(estaturas[estaturas > 175]))
print("Varianza    :", np.var(estaturas))
print("Percentil 75:", np.quantile(estaturas, 0.75))
````

**Control de valores en arrays**

Si deseamos asignar valores de acuerdo al cumplimiento de condiciones para un vector (similar a `ifelse` y `case_when` en R), las funciones `np.where` (ifelse) y `np.select` (case_when) son la solución, las cuales funcionan como:

```python
np.where(condicion, que hacer si es cierto, que hacer si es falso)
np.select(lista condiciones, lista elecciones, en otro caso)
```

````python
notas = np.array([2.5, 7, 5.2, 1.4, 6.8, 5, 4.9, 5.8])
notas
````

````python
# np.where
estado_aprobacion = np.where(notas >= 4, "Aprobado", "Reprobado")
estado_aprobacion
````

````python
# np.select
estado_examen = np.select(
    # Condiciones
    condlist=[notas > 5.5, notas >= 4.0],
    # Elecciones
    choicelist=["Eximido", "A dar el examen"],
    # Else
    default="A salvar el ramo",
)

estado_examen
````
